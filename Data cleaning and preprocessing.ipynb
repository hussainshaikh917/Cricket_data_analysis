# Data Cleaning and Preprocessing

Importing important libraries

```python
import pandas as pd
import json
```

### Converting Match Results JSON into CSV

```python
with open('C:/Users/Aniruddha Shelke/Documents/Projects/T20 Cricket WC Analysis/t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

data
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])

df_match.head(10)
```

```python
df_match.shape
```

Renaming "scorecard" column as "match_id" 

```python
df_match.rename({ 'scorecard' : 'match_id'}, axis = 1, inplace = True)

df_match.columns
```

```python
df_match.dtypes
```

```python
df_match.head()
```

Creating a match ids dictionary that maps team names to a unique match id to link with other tables

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv("C:/Users/Aniruddha Shelke/Documents/Projects/T20 Cricket WC Analysis/t20_csv_files/match_summary.csv", index = False)
```

### Converting Batting Summary JSON into CSV

```python
with open('C:/Users/Aniruddha Shelke/Documents/Projects/T20 Cricket WC Analysis/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)

all_results = []
for rec in data:
    all_results.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_results)

df_batting.tail()
```

```python
df_batting.shape
```

Adding new column "out/not_out" based on dismissal and then deleted dismissal column

```python
df_batting["out/not_out"] = df_batting["dismissal"].apply( lambda x : "out" if len(x) > 0 else "not_out")

df_batting.head(15)
```

```python
df_batting.drop(columns = ["dismissal"], inplace = True)

df_batting.head()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))

df_batting.head(10)
```

Adding and linking match_id column

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

Removing special characters from names

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(20)
```

```python
df_batting.shape
```

```python
df_batting.to_csv("C:/Users/Aniruddha Shelke/Documents/Projects/T20 Cricket WC Analysis/t20_csv_files/batting_summary.csv", index = False)
```

### Converting Player info JSON to CSV

```python
with open('t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)

df_players.head(10)
```

```python
df_players.shape
```

Cleaning up Special Characters from names

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))

df_players.head(10)
```

```python
df_players.to_csv("C:/Users/Aniruddha Shelke/Documents/Projects/T20 Cricket WC Analysis/t20_csv_files/player_summary.csv", index = False)
```
